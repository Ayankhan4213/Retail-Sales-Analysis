```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_data.csv")
df
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.shape
```

```python
df.describe()
```

```python
Data Overview

Dataset contains 10 rows and 6 columns.

Columns include OrderID, Date, Product, Category, Quantity, and Price.

Quantity and Price are numeric fields.

No missing values detected.
```

```python
df["Revenue"]=df["Quantity"]*df["Price"]
df
```

```python
total_revenue = df["Revenue"].sum()
total_revenue
```

```python
💰 Total Revenue
The total revenue generated from all sales transactions is ₹ _424800_____.
    
```

```python
df.groupby("Product")["Revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Category")["Revenue"].sum()
```

```python
📦 Category Performance

Electronics category contributes the majority of total revenue compared to Accessories.
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month
```

```python
df
```

```python
monthly_sales = df.groupby("Month")["Revenue"].sum()
monthly_sales
```

```python
📈 Monthly Trend

Sales increased/decreased across months showing seasonal variation in customer purchases.
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Product"], df["Revenue"])
plt.xlabel("UTPAAD")
plt.ylabel("RAJASVA")
plt.title("Revenue by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
monthly_sales.plot(kind="line", marker="o")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
📝 Conclusion

This project analyzed retail sales data using Python and Pandas.

A new revenue feature was created to measure business performance.

Electronics products generated the highest revenue.

Monthly trend analysis showed variation in customer purchasing behavior.

Visualizations helped identify top-performing products and sales trends.

This analysis can help business managers improve inventory and sales strategy.
```
